```python
from nltk.corpus import brown

word_list = brown.words()
five_letter_words = list(
    set([word.lower() for word in word_list if len(word) == 5 and word.isalpha()])
)

print(len(five_letter_words))
```

```python
# remove all words that arent only letters
five_letter_words = [word for word in five_letter_words if word.isalpha()]
five_letter_words
```

```python
len(five_letter_words)
```

```python
second_list = []
bad_letters = [c for c in "rtuiopsdfghcnl"]
for word in five_letter_words:
    if not any(letter in word for letter in bad_letters):
        if "a" == word[0] and "e" == word[3]:
            second_list.append(word)
print(len(second_list))
```

```python
second_list
```

```python
bad_letters = bad_letters + ["m"]
third_list = []
for word in second_list:
    if not any(letter in word for letter in bad_letters):
        if "e" in word and word[1] != "s" and word[4] != "e":
            if "r" in word and word[2] != "r":
                if "i" in word and word[3] != "i":
                    if "t" in word and word[4] != "t":
                        third_list.append(word)

print(len(third_list))
print(third_list)
```

```python
# pick random word from word_list
import random

word = random.choice(five_letter_words)
word
```
