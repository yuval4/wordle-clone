```python
# utility file to get some stats on the data

import glob
import json
import matplotlib.pyplot as plt

data_dir = "../webapp/data/"

language_codes = [f.split("/")[-1] for f in glob.glob(f"{data_dir}/languages/*")]
print("Total languages:", len(language_codes))

# load the 5words.txt file in each language and count the number of words
for code in language_codes:
    with open(f"{data_dir}languages/{code}/{code}_5words.txt") as f:
        words = f.read().split("\n")
        print(f"{code}: {len(words)} words")
```

```python
# plot the number of words for each language
n_words = {}
for code in language_codes:
    with open(f"{data_dir}languages/{code}/{code}_5words.txt") as f:
        words = f.read().split("\n")
        n_words[code] = len(words)
    # append supplemental words in {code}_5words_supplement.txt if it exists
    try:
        with open(f"{data_dir}languages/{code}/{code}_5words_supplement.txt") as f:
            words = f.read().split("\n")
            n_words[code] += len(words)
    except FileNotFoundError:
        pass


# sort by number of words
sorted_list = []
for code in sorted(n_words, key=n_words.get, reverse=True):
    sorted_list.append((code, n_words[code]))


# plot
plt.figure(figsize=(20, 10))
plt.bar(range(len(sorted_list)), [n for _, n in sorted_list], align="center")
plt.xticks(range(len(sorted_list)), [code for code, _ in sorted_list], rotation=60)
plt.savefig("out/n_words.png")
plt.show()
```

```python
# print a status list for the README


def load_languages():
    """returns a dict of language codes mapped to their english name and native name"""

    languages = {}
    # for each language folder, get the language config.name and config.name_natove
    for lang in language_codes:
        with open(f"{data_dir}languages/{lang}/language_config.json", "r") as f:
            language_config = json.load(f)
        languages[lang] = {
            "language_name": language_config["name"],
            "language_name_native": language_config["name_native"],
            "language_code": lang,
        }
    return languages


languages = load_languages()


def load_supplemental_words(lang):
    try:
        with open(f"{data_dir}languages/{lang}/{lang}_5words_supplement.txt", "r") as f:
            supplemental_words = [line.strip() for line in f]
    except FileNotFoundError:
        supplemental_words = []
    return supplemental_words


# sorted list of language_codes
sorted_language_codes = [item[0] for item in sorted_list]

exceptions = ["en", "fi"]  # these have been manually vetted
with open("out/status_list.txt", "w", encoding="utf-8") as f:
    for code in sorted_language_codes:
        with open(f"{data_dir}languages/{code}/{code}_5words.txt") as g:
            words = g.read().split("\n")
            words_supplement = load_supplemental_words(code)

            status_emoji = "🟥"
            if len(words) + len(words_supplement) > 500:
                status_emoji = "🟧"
            if len(words) + len(words_supplement) > 1000:
                status_emoji = "🟨"
            # if we've added words to the supplemental list, then we probably have a pretty good
            # "good words" + "possible" words list, thus green status emoji
            if len(words_supplement) > 2 or code in exceptions:
                status_emoji = "🟩"

            language_name = languages[code]["language_name"]
            f.write(
                f"  - {language_name} ({code}): {' '*(25-(len(language_name)+len(code)))} {status_emoji} ({len(words) + len(words_supplement)} words)\n"
            )
```
